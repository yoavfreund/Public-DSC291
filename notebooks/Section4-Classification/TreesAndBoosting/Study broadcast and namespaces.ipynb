```python
from pyspark import SparkContext
sc=SparkContext()
```

```python
class simple:
    def __init__(self):
        self.RDD=sc.parallelize(list(range(2)))
        self.m='initial'
    def Map(self):
        g='gg'
        B=sc.broadcast(self.m)
        def f(A):
            return (A,B.value,g)
        return self.RDD.map(f)
    def iterMap(self):   # the broadcast variable is available when calling the function 
                         # directly
        B=sc.broadcast(213)
        g='kk'
        def f(A):
            return (A,B.value,g)
        for A in self.RDD.collect():
            print(f(A))
```

```python
S=simple()
```

```python
S.Map().collect()
```
