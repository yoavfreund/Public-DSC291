```python
import pickle as pk
# with open('from_remote/worker-0/multi_processing_stats.pkl','rb') as pkl_file:
with open('multi_processing_stats.pkl','rb') as pkl_file:
    stat=pk.load(pkl_file)
stat
```

```python
%pylab inline
```

```python
figure(figsize=[10,10])
Ts=[stat['Times_compute'],stat['Times_read']]
_legend=['compute','read']
for i in [0,1]:
    Times=Ts[i]
    N=arange(0,len(Times))
    Times=array(Times)
    plot(N,Times,label=_legend[i])
grid()
xlabel('No. of threads')
ylabel('seconds')
legend()
```
